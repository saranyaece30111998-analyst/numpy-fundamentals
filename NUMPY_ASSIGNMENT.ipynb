# Numpy Array Operations:

```python
# importing numpy
import numpy as np
```

```python
#1. Create a 1D NumPy Array:
temperatures_w1=np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])
```

# 2. Inspection and Properties:

```python
# shape of array
print(" shape of array")
print(temperatures_w1.shape)

# data type of array
print("data type of array")
print(temperatures_w1.dtype)

#find the number of element
print("number of elements")
print(temperatures_w1.size)
```

# 3. Array Operations:

```python
# convert celsius in fahrenheit
Fahrenheit = (temperatures_w1* 9/5) + 32
print("temperature in fahrenheit")
print(Fahrenheit)


# maximum temperature
print("maximum temperature")
print(temperatures_w1.max())


# minimum temperature
print("minimum temperature")
print(temperatures_w1.min())

# mean temperature
print("mean temperature")
print(temperatures_w1.mean())
```

# 4. Array Slicing and Indexing:

```python
# Extract the temperatures for the first three days of the week.
print("temperatures of three days")
print(temperatures_w1[0:3])

#Extract the temperatures for the weekend (last two days).
print("temperatures of weekend")
print(temperatures_w1[5:7])


#● Extract the temperatures for the middle three days of the week.
print("temperatures of middle three days")
print( temperatures_w1[2:5])
```

# **PANDAS SERIES**

```python
# crating pandas seies
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['rank1', 'rank2', 'rank3', 'rank4', 'rank5']
)
print(marks)
```

#2. Indexing and Slicing:

```python
#Use the integer index position to access the mark of the 1st rank student.
print("mark of 1st rank student")
print(marks.iloc[0])

# mark of first three ranks
print("mark of first three ranks")
print(marks.loc[['rank1', 'rank2', 'rank3']])

#iloc accessor to retrieve the mark of the 3rd-ranked student.
print("mark of 3rd rank student")
print(marks.iloc[2])

# boolean mask to filter marks > 90
print("student with mark above 90")
print(marks[marks > 90])
```

# 3. Manipulating Series:

```python
# 1. modify the mark of the student with 1st rank to 100
marks['rank1'] = 100
print("mark of first rank")
print(marks['rank1'] )
```

```python
# 2. remove the entry corresponding to the last rank
marks = marks.drop('rank5')
```

```python
# ● Compute the CGPA by dividing each mark in the Series by 10.
cgpa=marks/10
print("cgpa of students is ")
print(cgpa)
```

# **Pandas DataFrame:**

# 1. Creating Pandas DataFrame:

```python
import pandas as pd

# create the dataframe
transactions = pd.DataFrame({
    'transactionid': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'productcategory': ['electronics', 'clothing', 'electronics', 'furniture', 'clothing',
                        'electronics', 'furniture', 'clothing', 'furniture', 'electronics'],
    'region': ['north', 'south', 'north', 'east', 'west', 'north', 'east', 'west', 'south', 'north'],
    'amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
})

# display the dataframe
print(transactions)


```

# 2. Data Exploration:

```python
#displaying first five rows
print("first five rows is")
print(transactions.head())

# displaying last five rows
print("last five rows is")
print(transactions.tail())

# shape of data frames
print("shape of data frames")
print(transactions.shape)

# displaying columns
print("columns of data frame ")
print(transactions.columns)

#printing data tpes
print("data type of data frame")
print(transactions.dtypes)
```

```python
# display only productcategory and amount
print(transactions[['productcategory', 'amount']])

# retrieve last 3 columns using iloc
print(transactions.iloc[:, -3:])

# filter rows where region is 'north' and amount > 200
print(transactions[(transactions['region'] == 'north') & (transactions['amount'] > 200)])

# value counts for productcategory
print(transactions['productcategory'].value_counts())

# unique values in region
print(transactions['region'].unique())

# group by region and find mean amount
print(transactions.groupby('region')['amount'].mean())
```

# 3. Manipulating the DataFrame:

```python
# modify amount for transactionid 102 to 165
transactions.loc[transactions['transactionid'] == 102, 'amount'] = 165

# add new column discount (10% of amount)
transactions['discount'] = transactions['amount'] * 0.10

# remove row with transactionid 109
transactions = transactions[transactions['transactionid'] != 109]

# delete the discount column
transactions = transactions.drop(columns=['discount'])
```
