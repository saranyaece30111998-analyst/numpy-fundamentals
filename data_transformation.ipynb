# **Data Cleaning**





```python
#drop a single column
df.drop('Useless_Col', axis=1, inplace=True)
```

```python
#drop a row (by index)
df.drop(204, axis=0, inplace=True)
```

# 2.Data quality checks

```python
#Display Duplicate Rows
df[df.duplicated()]
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

# name column cleaning

```python
df['Name']=df['Name'].str.title().str.strip()
```

# AGE COLUMN



```python
df['Age'].fillna(df['Age'].mean,inplace=True)
```

DEPARTMENT COLUMN

```python
# finf unique value in department
df['Department'].unique()
```

```python
# find value count of department
df['Department'].value_counts()
```

```python
# upper casing all names
df['Department']=df['Department'].str.upper()
```

```python
# replace values
df['Department'].replace('FINANACE','Finance',inplace=True)
```

```python
# replace multiple values
df['Department'].replace({'FINANCE':'Finance',
                          'MARKETING':'Marketing',
                          'SALES':'Sales'},inplace=True)
```

# position column

```python
df['Position'].unique()
```

```python
# Remove any special characters like !, @, #, etc.
df['Position']=df['Position'].str.replace(r'[^a-zA-Z ]',' ', regex=True)
```

# salary column

```python
df['Salary'].unique()
```

```python
#DISPLAYING SALARY
df["Salary"].value_counts(dropna = False)
```

```python
df["Salary"] = df['Salary'].str. replace("70k","70000").str.replace(",","")


```

```python
# changeing data types
df["Salary"]= df["Salary"].astype(float)
```

```python
# Hierarchical imputation
# Level 1: Department + Position mean
dept_pos_mean = df.groupby(['Department','Position'])['Salary'].transform('median')

# # # # Level 2: Department mean
dept_mean = df.groupby('Department')['Salary'].transform('mean')

# # # Level 3: Overall mean
overall_mean = df['Salary'].mean()

# Hierarchical imputation
df['Salary'] = (df['Salary'].fillna(dept_pos_mean).fillna(dept_mean).fillna(overall_mean))
```

# Feedback column

```python
df["Feedback"].unique()
```

```python
df["Feedback"].value_counts()
```

```python
df["Feedback"] = df['Feedback'].str. replace("?","Average").str.title()
```

```python
df["Feedback"] = df['Feedback'].fillna("Average")
```

```python
df["Has_Car"].unique()
```

```python
df['Has_Car'].replace({
    '0': 'No',
    '1': 'Yes'
}, inplace=True)
```

```python
df["Has_Car"] = df['Has_Car'].fillna("No")
```
