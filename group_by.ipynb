```python
import pandas as pd
```

```python
# Create first DataFrame (Employee details)
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5, 6],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Age': [25, 30, 28, 35, 26, 40],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Marketing', 'Finance'],
    'Salary': [45000, 60000, 52000, 75000, 48000, 82000]
})

# Create second DataFrame (Location details)
df2 = pd.DataFrame({
    'ID': [1, 2, 3, 4, 6, 7],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Chennai', 'Hyderabad', 'Pune'],
    'State': ['Delhi', 'Maharashtra', 'Karnataka', 'Tamil Nadu', 'Telangana', 'Maharashtra'],
    'Country': ['India', 'India', 'India', 'India', 'India', 'India']
})
print(df1)
print(df2)
```

```python
df_inner = pd.merge(
    df1[['ID', 'Name', 'Salary']],
    df2[['ID', 'City']],
    on='ID',
    how='inner'
)
df_inner
```

```python
df_left = pd.merge(
    df1[['ID', 'Name', 'Salary']],
    df2[['ID', 'City']],
    on='ID',
    how='left'
)
df_left
```

```python
df_right = pd.merge(
    df1[['ID', 'Name', 'Salary']],
    df2[['ID', 'City']],
    on='ID',
    how='right'
)
df_right
```

```python
#sample data
data = {
    'Region': ['East', 'West', 'East', 'North', 'West', 'East'],
    'Sales': [200, 300, 250, 400, 150, 350],
    'Profit': [20, 30, 25, 40, 15, 35]
}

df =pd.DataFrame(data)
```

```python
# Find total sales generated in each region
region_sales=df.groupby("Region")["Profit"].sum()
print(region_sales)
```

```python
#Calculate average profit for each region
region_average_profit=df.groupby("Region")["Profit"].mean()
print(region_average_profit)
```
