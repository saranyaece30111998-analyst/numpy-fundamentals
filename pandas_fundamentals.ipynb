```python
# Creating Dictionary
data = {
    'ID': [101, 102, 103, 104, 105, 106],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Age': [24, 30, 28, 35, 22, 40],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Berlin', 'Sydney'],
    'Salary': [50000, 60000, 55000, 70000, 48000, 75000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Marketing', 'Finance']
}

# Convert Dictionary to DataFrame
df = pd.DataFrame(data)

print("Original DataFrame")
print(df)
```

```python
df
```

```python
## Accesssing columns in pandas DataFrame
df["City"]
```

```python
# one or Columns using
print(df[["Name","City"]])
```
