# BASIC STATISTICAL AND MATHEMATICAL OPERATIONS

```python
import numpy as np
# Create a 2D NumPy array with 2 rows and 3 columns
arr =np.array([[6,2,3],
              [4,5,1]])
```

```python
# Minimum value in the array and index
arr.min()
print(np.argmin(arr))
```

```python
# Maximum value in the array
print(arr.max())
print(np.argmax(arr))
```

```python
# Sum of all elements
print(arr.sum())
```

```python
# Product of all elements
print(np.prod(arr))
```

```python
# Mean / Average of the array
print(arr.mean())
print(arr.var())
print(arr.std())
```

NUMPY RANDOM FUNCTIONS

```python
# 1. np.random.rand(d1, d2,..)
# Uniform random numbers in(0,1)

np.random.rand(5,2)
```

```python
# 2. np.random.randn(d1, d2, ...)
# Standard normal distribution (mean=0, std=1)

np.random.randn(5,2)
```

```python
# 3. np.random.randint(low, high, size)
#Random integers in a range
np.random.randint(2,4,30)
```

```python
# 4. np.random.choice(arr, size)
#Random sample from an existing array
np.random.seed(42)
arr =np.array([2,3,4,5,6])
np.random.choice(arr,20)
```

```python
# 5. np.random.seed(n)
#Set seed for reproducibility
np.random.seed(42)
```
